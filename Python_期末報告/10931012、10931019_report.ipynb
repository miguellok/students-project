```python
import matplotlib.pyplot as plt
```

```python
#labe → 圖例
listx=[1,5,7,9,13,16]
listy=[15,50,80,40,70,50]
plt.plot(listx,listy,'o-.',color='red',label='money')
plt.legend()
plt.show()
```

```python
#若沒有設定顏色，系統會自動設定不同顏色
listx1=[1,5,7,9,13,16]
listy1=[15,50,80,40,70,50]
plt.plot(listx1,listy1,label='Male')
listx2=[2,6,8,11,14,16]
listy2=[10,40,30,50,80,60]
plt.plot(listx2,listy2,label='Female')
plt.legend()#label屬性設定後，需執行legend才會顯示
plt.title('Pocket Money')
plt.xlabel('Age')
plt.ylabel('Money')
plt.xlim(0,20)#模組.xlim(起始值，終止值)#設定x座標範圍
plt.ylim(0,100)#模組.ylim(起始值，終止值)#設定y座標範圍
```

```python
#labeldistance → 項目標題與圓心的距離是半徑的多少倍
#autopct → 項目百分比的格式，語法為「%2.1f%%」表示整數2位數，小數1位數
#shadow → 布林值
#startangle → 開始繪圖的起始角度，繪圖會以逆時針旋轉計算角度
#axis → 若要繪製正圓形圓餅圖，須以axis('equal')讓x、y軸單位相等
labels=['東部','南部','北部','中部']
sizes=[5,10,20,15]
colors=['red','orange','pink','yellow']#每一個顏色組成組成的串列
explode=(0,0,0.5,0)#每一個項目凸出數值組成的串列，0為未凸出，數字越大凸出越明顯
plt.pie(sizes,explode=explode,labels=labels,colors=colors,labeldistance=1.1,autopct='%3.1f%%',shadow=True,startangle=90,pctdistance=0.6)
plt.axis('equal')
plt.title('比例圖')
plt.legend()
plt.show()
```

```python
label2=['東部','南部','北部','中部']
size1=[5,10,20,15]
color=['pink','lightcoral','yellow','linen']
plt.pie(size1,colors=color,labels=label2,autopct='%1.1f%%')
```

```python
datas2=[Scatter(x=['林大名','陳曉明','蔡小美','廉傑克曼'],y=[67,89,72,95])]
plotly.offline.iplot({'data':datas2})
```

## Plotly繪圖進階設定

### mode='mode屬性值'
### 當mode屬性值為'markers'時畫出散點圖；當mode屬性值為'lines'時畫出線形圖

```python
datas3=[Scatter(x=['林大名','陳曉明','蔡小美','廉傑克曼'],y=[67,89,72,95],mode='markers')]
plotly.offline.iplot({'data':datas3})
```

### 為繪圖點加入文字

```python
datas4=[Scatter(x=['林大名','陳曉明','蔡小美','廉傑克曼'],y=[67,89,72,95],text=['1號','2號','3號','4號'],textposition='top center',mode='markers+lines+text')]
plotly.offline.iplot({'data':datas4})
```

## 圖形標題
### 'layout':Layout(title='標題名稱')

```python
from plotly.graph_objs import Scatter,Layout
datas5=[Scatter(x=['林大名','陳曉明','蔡小美','廉傑克曼'],y=[67,89,72,95],text=['1號','2號','3號','4號'],textposition='top center',mode='markers+lines+text')]
plotly.offline.iplot({'data':datas5,'layout':Layout(title='二年甲班成績單')})
```

```python
datas6=[Scatter(x=['林大名','陳曉明','蔡小美','廉傑克曼'],y=[67,89,72,95],name='國文'),Scatter(x=['林大名','陳曉明','蔡小美','廉傑克曼'],y=[72,57,96,77],name='英文'),Scatter(x=['林大名','陳曉明','蔡小美','廉傑克曼'],y=[83,87,68,59],name='數學')]
plotly.offline.iplot({'data':datas6,'layout':Layout(title='二年甲班成績單')})
```

### 利用Candlestick繪製股票K線圖，以Apple公司為例
### 綠色為漲(開盤價<收盤價)；紅色為跌(開盤價>收盤價)

```python
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv')
fig = go.Figure(data=[go.Candlestick(x=df['Date'],open=df['AAPL.Open'], high=df['AAPL.High'],low=df['AAPL.Low'], close=df['AAPL.Close'])])
iplot(fig)
```

## Array
### ◆ 多數據處理的指令
### ◆ 每一個元素都有相同的資料型態

```python
import numpy as np
plt.figure(figsize=(13,7))
plt.bar(covid['日期'],covid['本土個案'],color='royalblue',label='本土個案')
plt.bar(covid['日期'],covid['校正回歸'],color='orange',bottom=np.array(covid['本土個案']),label='校正回歸')

plt.ylim(0,800)
plt.xlabel('日期',fontsize=15)
plt.ylabel('確診人數',fontsize=15)
plt.xticks(fontsize=11,rotation=45)
plt.yticks(fontsize=12)
plt.title('2021年五、六月確診人數',fontsize=20)
plt.legend()
```

```python
import plotly
from plotly.graph_objs import Scatter,Layout
plotly.offline.init_notebook_mode(connected=True) 
data=[Scatter(x=covid['日期'],y=covid['本土個案'],name='本土個案'),Scatter(x=covid['日期'],y=covid['校正回歸'],name='校正回歸'),Scatter(x=covid['日期'],y=covid['死亡人數'],name='死亡人數'),Scatter(x=covid['日期'],y=covid['境外移入'],name='境外移入'),Scatter(x=covid['日期'],y=covid['當日總計'],name='當日總計')]
plotly.offline.iplot({'data':data,'layout':Layout(title='2021年五、六月確診人數')})
```

### 疊加 : Layout(barmode='stack') 

```python
import plotly
from plotly.graph_objs import Bar,Scatter,Layout
plotly.offline.init_notebook_mode(connected=True) 
data1=[Bar(x=covid['日期'],y=covid['本土個案'],name='本土個案'),Bar(x=covid['日期'],y=covid['校正回歸'],name='校正回歸'),Scatter(x=covid['日期'],y=covid['死亡人數'],name='死亡人數',)]
plotly.offline.iplot({'data':data1,'layout':Layout(barmode='stack',title='2021年五、六月確診人數')})
```
